```python
from anthropic import Anthropic

def estimate_anthropic_tokens(text: str) -> int:
    client = Anthropic()
    return client.count_tokens(text)

def get_exact_token_count(text: str, model: str = "claude-3-sonnet-20240229") -> int:
    client = Anthropic()
    response = client.messages.create(
        model=model,
        max_tokens=1,
        messages=[{"role": "user", "content": text}]
    )
    return response.usage.input_tokens

# Example usage
estimated_count = estimate_anthropic_tokens(full_text)
print(f"Estimated token count: {estimated_count}")

exact_count = get_exact_token_count(full_text)
print(f"Exact token count: {exact_count}")
```

```python
import llmsherpa
    # ... existing code ...

def to_markdown(doc):
    """
    Converts the document to a markdown formatted string.
    """
    markdown = ""
    
    for section in doc.sections():
        # Add section header
        markdown += f"{'#' * (section.level + 1)} {section.to_text()}\n\n"
        
        for child in section.children:
            if isinstance(child, llmsherpa.readers.layout_reader.Paragraph):
                markdown += f"{child.to_text()}\n\n"
            elif isinstance(child, llmsherpa.readers.layout_reader.ListItem):
                markdown += f"- {child.to_text()}\n"
            elif isinstance(child, llmsherpa.readers.layout_reader.Table):
                markdown += child.to_text() + "\n\n"
    
    return markdown.strip()
```

```python
import re
from typing import List, Dict, Tuple
from difflib import SequenceMatcher

from plotreader.utils.base import BasicAnthropicLLM

class MarkdownMerger:
    def __init__(self, version1: str, version2: str):
        self.version1 = version1
        self.version2 = version2

    def parse_markdown(self, markdown: str) -> List[Tuple[int, str, str]]:
        """
        Parse markdown into a list of (level, header, content) tuples.
        """
        sections = []
        current_level = 0
        current_header = ""
        current_content = []

        for line in markdown.split('\n'):
            if line.startswith('#'):
                if current_header:
                    sections.append((current_level, current_header, '\n'.join(current_content)))
                current_level = len(line.split()[0])  # Count the number of '#'
                current_header = line.lstrip('#').strip()
                current_content = []
            else:
                current_content.append(line)

        if current_header:
            sections.append((current_level, current_header, '\n'.join(current_content)))

        return sections

    def similarity_score(self, a: str, b: str) -> float:
        return SequenceMatcher(None, a, b).ratio()

    def match_sections(self, sections1: List[Tuple[int, str, str]], 
                       sections2: List[Tuple[int, str, str]]) -> List[Tuple[Tuple[int, str, str], Tuple[int, str, str]]]:
        """
        Match sections based on header similarity.
        """
        matched = []
        for s1 in sections1:
            best_match = max(sections2, key=lambda s2: self.similarity_score(s1[1], s2[1]))
            if self.similarity_score(s1[1], best_match[1]) > 0.6:  # Threshold for considering a match
                matched.append((s1, best_match))
        return matched
    
    def merge_content(self, header: str,content1: str, content2: str) -> str:
        """
        Merge content from two versions.
        """
        try:
            return self._merge_content_llm(header, content1, content2)
        except Exception as e:
            print(e)
            return self._merge_content_length_based(header, content1, content2)
    
    def _merge_content_length_based(self, header: str,content1: str, content2: str) -> str:
        """
        Merge content from two versions, preferring longer paragraphs.
        """

        paragraphs1 = content1.split('\n\n')
        paragraphs2 = content2.split('\n\n')
        
        merged = []
        for p1, p2 in zip(paragraphs1, paragraphs2):
            merged.append(p1 if len(p1) >= len(p2) else p2)
        
        # Add any remaining paragraphs from the longer version
        merged.extend(paragraphs1[len(paragraphs2):])
        merged.extend(paragraphs2[len(paragraphs1):])
        
        return '\n\n'.join(merged)

    def _merge_content_llm(self, header: str,content1: str, content2: str) -> str:
        """
        Merge content from two versions, using an LLM.
        """
        

        claude = BasicAnthropicLLM(model="claude-3-haiku-20240307")

        prompt = f"""Below are two different markdown strings extraced from a scientific paper via different parsers.
        Your job is to merge the two markdown strings into a single markdown string that is your best estimate of the original text.
        The goal would be for each figure caption to be its own section.
        
        SECTION HEADER TITLE:
        {header}

        PARSED CONTENT 1:
        {content1}

        PARSED CONTENT 2:
        {content2}

        Return only the final markdown string without any other text.
        """

        response = claude.message(prompt)
        raw_text = response.content[0].text
        split_text = raw_text.split("\n\n")
        header = split_text[0]
        header = header.replace("#", "").strip()
        content = "\n\n".join(split_text[1:])
        return header, content

    def merge(self) -> str:
        """
        Perform the entire merge process and return the final markdown.
        """
        sections1 = self.parse_markdown(self.version1)
        sections2 = self.parse_markdown(self.version2)
        
        matched_sections = self.match_sections(sections1, sections2)
        
        merged_sections = []
        for s1, s2 in matched_sections:
            level = min(s1[0], s2[0])  # Use the lower level (higher in hierarchy)
            header = s1[1] if len(s1[1]) >= len(s2[1]) else s2[1]
            header, content = self.merge_content(f"{'#' * level} {header}", s1[2], s2[2])
            merged_sections.append((level, header, content))
        
        # Sort merged sections based on their order in version1
        merged_sections.sort(key=lambda x: sections1.index(x[0]) if x[0] in sections1 else len(sections1))
        
        # Generate final markdown
        final_markdown = ""
        for level, header, content in merged_sections:
            final_markdown += f"{'#' * level} {header}\n\n{content}\n\n"
        
        return final_markdown.strip()

# Usage
# merger = MarkdownMerger(version1, version2)
# final_markdown = merger.merge()
```

```python
from typing import List, Dict, Tuple
from difflib import SequenceMatcher
import re
from llama_index.core import Settings
from llama_index.llms.anthropic import Anthropic

class MarkdownNode:
    def __init__(self, level: int, header: str, content: str = ""):
        self.level = level
        self.header = header
        self.content = content
        self.children = []

    def add_child(self, child: 'MarkdownNode'):
        self.children.append(child)

class MarkdownMerger:
    def __init__(self, version1: str, version2: str):
        self.version1 = version1
        self.version2 = version2
        self.llm = Settings.llm or Anthropic(model="claude-3-haiku-20240307")

    def parse_markdown_to_tree(self, markdown: str) -> List[MarkdownNode]:
        lines = markdown.split('\n')
        root = MarkdownNode(0, "Root")
        stack = [root]
        current_content = []

        for line in lines:
            if line.startswith('#'):
                if current_content:
                    stack[-1].content = '\n'.join(current_content)
                    current_content = []

                level = len(line.split()[0])
                header = line.lstrip('#').strip()
                node = MarkdownNode(level, header)

                while stack[-1].level >= level:
                    stack.pop()
                
                stack[-1].add_child(node)
                stack.append(node)
            else:
                current_content.append(line)

        if current_content:
            stack[-1].content = '\n'.join(current_content)

        return root.children

    def text_similarity(self, text1: str, text2: str) -> float:
        return SequenceMatcher(None, text1, text2).ratio()

    def find_best_match(self, node: MarkdownNode, candidates: List[MarkdownNode]) -> Tuple[MarkdownNode, float]:
        best_match = None
        best_score = 0

        for candidate in candidates:
            node_text = node.header + "\n\n" + node.content
            candidate_text = candidate.header + "\n\n" + candidate.content
            similarity_score = self.text_similarity(node_text, candidate_text)

            # if combined_score > 0.9:
            #     continue
            if similarity_score > best_score:
                best_score = similarity_score
                best_match = candidate

        return best_match, best_score

    def merge_trees(self, tree1: List[MarkdownNode], tree2: List[MarkdownNode]) -> List[MarkdownNode]:
        merged = []
        used2 = set()

        if len(tree1) == 0:
            if len(tree2) == 0:
                return []
            return tree2
        if len(tree2) == 0:
            return tree1

        for node1 in tree1:
            best_match, score = self.find_best_match(node1, tree2)
            print(score)
            if score > 0.2:  # Adjusted threshold
                merged_node = self.merge_nodes(node1, best_match)
                merged.append(merged_node)
                used2.add(best_match)
            else:
                merged.append(node1)

        merged = sorted(merged, key=lambda x: x.level)

        # Add remaining nodes from tree2
        unused = []
        for node2 in tree2:
            if node2 not in used2:
                unused.append(node2)

        if len(unused) == 0 or len(used2) == 0:
            return merged
        else:
            return self.merge_trees(merged, unused)


    def merge_nodes(self, node1: MarkdownNode, node2: MarkdownNode) -> MarkdownNode:
        merged_header, merged_content = self.merge_content(node1.header, node1.content, node2.content)
        merged_node = MarkdownNode(min(node1.level, node2.level), merged_header, merged_content)
        merged_node.children = self.merge_trees(node1.children, node2.children)
        return merged_node

    def merge_content(self, header: str, content1: str, content2: str) -> Tuple[str, str]:
        try:
            return self._merge_content_llm(header, content1, content2)
        except Exception as e:
            print(f"LLM-based merging failed: {e}")
            return header, self._merge_content_length_based(content1, content2)

    def _merge_content_llm(self, header: str, content1: str, content2: str) -> Tuple[str, str]:
        prompt = f"""Below are two different markdown strings extracted from a scientific paper via different parsers.
        Your job is to merge the two markdown strings into a single markdown string that is your best estimate of the original text.
        Keep your eye out for text that could be figure captions. The parsers struggle with these. You may need to add headers or recombine text to best represent figure captions.
        The goal would be for each figure caption to be its own section.
        
        SECTION HEADER TITLE:
        {header}

        PARSED CONTENT 1:
        {content1}

        PARSED CONTENT 2:
        {content2}

        Return only the final markdown string without any other text.
        """

        response = self.llm.complete(prompt)
        raw_text = response.text
        split_text = raw_text.split("\n\n")
        merged_header = split_text[0].replace("#", "").strip()
        merged_content = "\n\n".join(split_text[1:])
        return merged_header, merged_content

    def _merge_content_length_based(self, content1: str, content2: str) -> str:
        paragraphs1 = content1.split('\n\n')
        paragraphs2 = content2.split('\n\n')
        
        merged = []
        for p1, p2 in zip(paragraphs1, paragraphs2):
            merged.append(p1 if len(p1) >= len(p2) else p2)
        
        merged.extend(paragraphs1[len(paragraphs2):])
        merged.extend(paragraphs2[len(paragraphs1):])
        
        return '\n\n'.join(merged)

    def tree_to_markdown(self, nodes: List[MarkdownNode], level: int = 0) -> str:
        markdown = ""
        for node in nodes:
            markdown += f"{'#' * node.level} {node.header}\n\n{node.content}\n\n"
            if node.children:
                markdown += self.tree_to_markdown(node.children, level + 1)
        return markdown

    def merge(self) -> str:
        tree1 = self.parse_markdown_to_tree(self.version1)
        tree2 = self.parse_markdown_to_tree(self.version2)
        merged_tree = self.merge_trees(tree1, tree2)
        return self.tree_to_markdown(merged_tree)
```

```python
from PIL import Image
from pydantic import BaseModel, Field
from typing import Optional
from llama_index.core.prompts import PromptTemplate
from llama_index.core.output_parsers.pydantic import PydanticOutputParser
from llama_index.core.schema import TextNode
from llama_index.core.node_parser import MarkdownNodeParser

import plotreader

class Section(BaseModel):
    header: str
    level: int

class PageMetadata(BaseModel):
    page_number: int
    figure_count: int
    figure_names: List[str]
    section_headers: List[str]
    contains_fig_caption: bool = Field(..., description="Does any part of a figure caption exist on this page?")
    # contains_partial_fig_caption: bool = Field(..., description="Is there a figure caption that is partially contained in this page?")

output_parser = PydanticOutputParser(output_cls=PageMetadata)
prompt_page_metadata = PromptTemplate(
    "The image provided is a page from a scientific paper. Is there a figure in this image? If so, what is the figure's name?",
    output_parser=output_parser
).format(llm=plotreader._CLAUDE_SONNET35_MULTIMODAL)

# You've also been provided with an image of the page and the page before and after (if they exist).
revision_prompt = """Below is a markdown string parsed from a signle page of a scientific paper PDF. 
Your job is to revise the levels of each part of the extracted markdown string so it matches the PDF.
To aid you, I've provided the list of expected section headers for the page and information about whether or not to expecdt figure captions.
I've also included the previous page's headers to inform the levels of the target page.
Plese make each Figure caption its own section. Note that some figure captions may be split across two pages.
Also, do not change any text. Only change the levels and organization of the sections to better match the PDF image.

Page number: {page_number}

EXPECTED SECTION HEADERS:
{sections}

PREVIOUS PAGE HEADERS:
{previous_page_headers}

PARSED MARKDOWN:
{content1}

CONTAINS WHOLE OR PARTIAL FIGURE CAPTION:
{contains_fig_caption}

PREVIOUS PAGE CONTAINS WHOLE OR PARTIAL FIGURE CAPTION:
{previous_page_contains_fig_caption}


Return only the final markdown string without any other text.
"""

page_metadata_responses = []
prev_page_contains_fig_caption = False
prev_page_headers = []
for image_dict in image_dicts:
    if image_dict.get("type") == "full_page_screenshot":

        # get page metadata
        page_num = image_dict.get("page_number")
        page_img = Image.open(image_dict.get("path"))
        response = claude_mm.message(prompt_page_metadata, images=[page_img])
        page_metadata_responses.append(output_parser.parse(response.content[0].text))

        # revise page markdown
        try:
            page_md = json_objs[0]["pages"][page_num-1]["md"]
            revision_prompt_with_md = revision_prompt.format(
                page_number=page_num, 
                sections=page_metadata_responses[page_num-1].section_headers, 
                content1=page_md,
                previous_page_contains_fig_caption=prev_page_contains_fig_caption,
                contains_fig_caption=page_metadata_responses[page_num-1].contains_fig_caption,
                previous_page_headers=f"\n".join(prev_page_headers),
            )
            print(revision_prompt_with_md)
            response = claude_mm.message(revision_prompt_with_md, images=[page_img])
            revised_md = response.content[0].text
            json_objs[0]["pages"][page_num-1]["md_revised"] = revised_md

            prev_page_contains_fig_caption = page_metadata_responses[page_num-1].contains_fig_caption
            node = TextNode(text=revised_md)
            nodes = MarkdownNodeParser().get_nodes_from_node(node)
            for node in nodes:
                for key in node.metadata.keys():
                    if key.startswith("Header_"):
                        level = int(key.split("_")[1])
                        prev_page_headers.append("".join(["#"]*level) + " " + node.metadata.get(key))
        except Exception as e:
            print(e)    
            json_objs[0]["pages"][page_num-1]["md_revised"] = page_md

```
